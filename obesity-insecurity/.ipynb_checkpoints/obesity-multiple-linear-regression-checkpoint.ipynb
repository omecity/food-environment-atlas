```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
# read in the data
data = pd.read_csv('../data/state_data.csv')
obesity_data = data.drop('Unnamed: 0', axis = 1)
```

```python
# create a bar plot to see obesity rate in each state
sns.barplot(obesity_data, x="State", y="PCT_OBESE_ADULTS17").set_title("Adult Obesity Rate by State")
plt.gcf().set_size_inches(15,5)
```

```python
# split into trainging and testing data
obesity_train, obesity_test = train_test_split(obesity_data.copy(),
                                          random_state = 73098,
                                          shuffle = True,
                                          test_size = .2)

# export testing data for reproducability
obesity_test.to_csv('../data/state_test.csv')
```

```python
# uncomment the last command to see what the variables are
help = pd.read_excel('../data/2016 Data Cleaned.xlsx')
var_labels = help[['Variable Code', 'Variable Name']].head(25)
#var_labels
```

```python
# split features in store/facility features and assistance features
# focus on featurs given as a proportion of the population

pth_store_cols = ['GROCPTH16', 'SUPERCPTH16', 'CONVSPTH16', 'SPECSPTH16', 'RECFACPTH16']
pth_assist_cols = ['WICSPTH16', 'FFRPTH16', 'FSRPTH16', 'SNAP_PART_RATE16','PC_WIC_REDEMP16', 'REDEMP_WICS16', 'PCT_WICINFANTCHILD16', 'PCT_WICWOMEN16']
```

```python
# plot obesity vs store features
sns.pairplot(
                data=obesity_train,
                x_vars=pth_store_cols,
                y_vars=['PCT_OBESE_ADULTS17'],
                hue='SNAP_BBCE16'
            )

plt.show()
```

```python
# plot obesity vs assistance features
sns.pairplot(
                data=obesity_train,
                x_vars=pth_assist_cols,
                y_vars=['PCT_OBESE_ADULTS17'],
                hue='SNAP_BBCE16'
            )

plt.show()
```

```python
# create a correlation matrix
correlation_matrix = obesity_train[pth_store_cols + pth_assist_cols + ['PCT_OBESE_ADULTS17']].corr()

# display the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix of Selected Variables')
plt.tight_layout()
plt.show()
```

```python
# import sklearn packages
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
```

```python
# let's do some more in-depth feature selection

# we will use lasso regression

alphas = [0.000001,.00001,.0001,.001,.01,.1,.25,.5,.75,1,5,10,25,50,100,1000,10000]
features = pth_store_cols + pth_assist_cols

## make coefficient holder
coefs = np.zeros((len(alphas), len(features)))

## scale quantitative variables
obesity_train_array = obesity_train[features].to_numpy()

scaler = StandardScaler()
obesity_train_array_copy = obesity_train_array.copy()
scaler.fit(obesity_train_array_copy)
obesity_train_array_scaled = scaler.fit_transform(obesity_train_array_copy)


## Loop through alphas
for i in range(len(alphas)):
    ## make lasso model
    lasso = Lasso(alpha=alphas[i], max_iter=10000000)
    
    
    ## fit model
    lasso.fit(obesity_train_array_scaled, obesity_train.PCT_OBESE_ADULTS17.values)
    
    ## record coefs
    coefs[i,:] = lasso.coef_
```

```python
df = pd.DataFrame(data=coefs, index=alphas, columns=features)
# df
```

```python
df.ne(0).idxmin().sort_values(ascending=False)
```

```python
# It seems like there are 5 features that stand out much more than the others.
# These features are:
selected_feats = ['SUPERCPTH16', 'CONVSPTH16', 'SPECSPTH16', 'WICSPTH16', 'FSRPTH16']
selected_feats_indices = [3, 5, 7, 9, 13]
```

```python
from pandas.plotting import scatter_matrix
```

```python
# visualize pairwise relationships
scatter_matrix(obesity_train[selected_feats + ['PCT_OBESE_ADULTS17']], figsize=(10,10), alpha=.9)

plt.show()

# it seems that multiple linear regression should be fine
# no need to do polynomial regression
```

```python
# make a Kfold object with k = 4
kfold = KFold(4, shuffle = True, random_state = 8700)

# k=4 chosen so that the holdout sets are roughly the same size of our testing set
```

```python
# let's try three different models:
    # 1. A baseline model that predicts the obesity rate will be the average of all the training data
    # 2. A multiple linear regression model 
    # 3. A k-nearest-neighbors regression model

# make an array of zeros that will hold our mses
mses = np.zeros((3, 4))

## This keeps track of what split we are on
i = 0
# fill in what is missing in the for loop declaration
for i,(train_index, test_index) in enumerate(kfold.split(obesity_train)):
    # now we get the training splits and the holdout split
    # Training
    obesity_tt = obesity_train.iloc[train_index]
    
    # export training sets for reproducability
    obesity_tt.to_csv('../data/state_training_{}.csv'.format(i))
    
    # Holdout
    obesity_ho = obesity_train.iloc[test_index]

    # export holdout sets for reproducability
    obesity_ho.to_csv('../data/state_holdout_{}.csv'.format(i))
    
    
    ### This is Model 0 ###
    # take the mean PCT_OBESE_ADULTS17 from the training set
    # we need predictions for the entire holdout set
    pred0 = obesity_tt.PCT_OBESE_ADULTS17.mean() * np.ones(len(obesity_ho))
    

    model1 = LinearRegression()
    model2 = KNeighborsRegressor(n_neighbors = 5)

    # fit models on the training data, obesity_tt
   
    model1.fit(obesity_tt[selected_feats], obesity_tt.PCT_OBESE_ADULTS17.values)
    model2.fit(obesity_tt[selected_feats], obesity_tt.PCT_OBESE_ADULTS17.values)

    
    # get the prediction on holdout set
    pred1 = model1.predict(obesity_ho[selected_feats])
    pred2 = model2.predict(obesity_ho[selected_feats])

    
    ### Recording the MSES ###
    mses[0,i] = mean_squared_error(obesity_ho.PCT_OBESE_ADULTS17.values, pred0)
    mses[1,i] = mean_squared_error(obesity_ho.PCT_OBESE_ADULTS17.values, pred1)
    mses[2,i] = mean_squared_error(obesity_ho.PCT_OBESE_ADULTS17.values, pred2)
mses
```

```python
# This figure will compare the performance
plt.figure(figsize=(8,5))

plt.scatter(np.zeros(4), 
            mses[0,:], 
            s=60, 
            c='white',
            edgecolor='black',
            label="Single Split")
plt.scatter(np.ones(4), 
            mses[1,:], 
            s=60, 
            c='white',
            edgecolor='black')
plt.scatter(2*np.ones(4), 
            mses[2,:], 
            s=60, 
            c='white',
            edgecolor='black')

plt.scatter([0,1,2], 
            np.mean(mses, axis=1), 
            s=60, 
            c='r',
            marker='X',
            label="Mean")

plt.legend(fontsize=12)

plt.xticks([0,1,2],["Baseline", "Multiple Linear Regression", "KNN Regression"], fontsize=10)
plt.yticks(fontsize=10)

plt.ylabel("MSE", fontsize=12)

plt.show()
```

```python
# what is the mse on the entire training set for our baseline model?
pred0 = obesity_train.PCT_OBESE_ADULTS17.mean() * np.ones(len(obesity_train))
mse_pred0 = mean_squared_error(obesity_train.PCT_OBESE_ADULTS17.values, pred0)
mse_pred0
```

```python
# Multiple linear regression seems to be our best bet
# train the model on our entire training set

model = LinearRegression()
model.fit(obesity_train[selected_feats], obesity_train.PCT_OBESE_ADULTS17.values)
pred_on_training = model.predict(obesity_train[selected_feats])

mse_train = mean_squared_error(obesity_train.PCT_OBESE_ADULTS17.values, pred_on_training)
mse_train
```

```python
# Find description of the selected features and reindex
selected_feats_info = pd.DataFrame(var_labels.loc[selected_feats_indices, 'Variable Name'])
selected_feats_info.set_index(np.arange(0,5), inplace = True)

# Print dataframe with feature code & description alongside its coefficient
# For interpretability
coefficients = pd.concat([pd.DataFrame(selected_feats), selected_feats_info, pd.DataFrame(model.coef_)], axis = 1)
coefficients.columns = ['Variable Code', 'Variable Name', 'Coefficient']
coefficients
```

```python
model.intercept_
```

```python
# compute residuals
residuals = obesity_train.PCT_OBESE_ADULTS17.values - pred_on_training
```

```python
# plot residuals vs the selected features

num_feats = 5

fig,ax = plt.subplots(1,num_feats,figsize=(20,2))

for i in range(0, num_feats):
    ax[i].scatter(obesity_train[selected_feats[i]], residuals)
    ax[i].set_xlabel(selected_feats[i], fontsize=10)
ax[0].set_ylabel("residuals", fontsize=10)
plt.show()

# observe the heteroskedasticity in the fourth plot - WICSPTH16
# otherwise these look good. Yay!
```

```python
# visualize our model on training data
# for each feature, project our model onto the PCT_OBESE_ADULT17 x feature plane

fig,ax = plt.subplots(1,num_feats,figsize=(20,2))

for i in range(0, num_feats):
    ax[i].scatter(obesity_train[selected_feats[i]], obesity_train["PCT_OBESE_ADULTS17"])
    ax[i].set_xlabel(selected_feats[i], fontsize=10)
    ax[i].plot([np.min(obesity_train[selected_feats[i]]), np.max(obesity_train[selected_feats[i]])], [model.intercept_, model.intercept_ + model.coef_[i]*(np.max(obesity_train[selected_feats[i]]))],
              'g-')
ax[0].set_ylabel("PCT_OBESE_ADULTS17", fontsize=10)
plt.show()
```

```python
# final sanity check on testing data

preds = model.predict(obesity_test[selected_feats])

mse = mean_squared_error(obesity_test.PCT_OBESE_ADULTS17.values, preds)
mse
```

```python
# what were the predictions and what are the actual values?
states = pd.DataFrame(obesity_test["State"]).set_index(np.arange(0,11))
rates = pd.DataFrame(obesity_test["PCT_OBESE_ADULTS17"]).set_index(np.arange(0,11))
compare = pd.concat([states, pd.DataFrame(preds), rates], axis = 1)
compare.columns = ["State", "Predicted Obesity Rate", "True Obesity Rate"]
compare
```
