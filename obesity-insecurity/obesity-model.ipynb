```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
# read in the data
data = pd.read_csv('./state_data.csv')
obesity_data = data.drop('Unnamed: 0', axis = 1)
```

```python
# split into trainging and testing data
obesity_train, obesity_test = train_test_split(obesity_data.copy(),
                                          random_state = 73098,
                                          shuffle = True,
                                          test_size = .2)
```

```python
# uncomment the last command to see what the variables are
help = pd.read_excel('../data/2016 Data Cleaned.xlsx')
var_labels = help[['Variable Code', 'Variable Name']].head(25)
#var_labels
```

```python
# split features in store/facility features and assistance features
pth_store_cols = ['GROCPTH16', 'SUPERCPTH16', 'CONVSPTH16', 'SPECSPTH16', 'RECFACPTH16']
pth_assist_cols = ['WICSPTH16', 'FFRPTH16', 'FSRPTH16', 'SNAP_PART_RATE16','PC_WIC_REDEMP16', 'REDEMP_WICS16', 'PCT_WICINFANTCHILD16', 'PCT_WICWOMEN16']
```

```python
# plot obesity vs store features
sns.pairplot(
                data=obesity_train,
                x_vars=pth_store_cols,
                y_vars=['PCT_OBESE_ADULTS17'],
                hue='SNAP_BBCE16'
            )

plt.show()
```

```python
# plot obesity vs assistance features
sns.pairplot(
                data=obesity_train,
                x_vars=pth_assist_cols,
                y_vars=['PCT_OBESE_ADULTS17'],
                hue='SNAP_BBCE16'
            )

plt.show()
```
