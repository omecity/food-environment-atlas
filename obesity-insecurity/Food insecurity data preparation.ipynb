```python
# Define the columns to replace null values with mean
columns_to_fill = ['LACCESS_POP15','LACCESS_LOWI15','LACCESS_HHNV15','LACCESS_SNAP15','LACCESS_CHILD15',
                   'PCT_LACCESS_CHILD15','LACCESS_SENIORS15','LACCESS_WHITE15','LACCESS_BLACK15',
                   'LACCESS_HISP15','LACCESS_NHASIAN15','LACCESS_NHNA15','LACCESS_MULTIR15','SNAPS17','WICS11','WICS16','MEDHHINC15','POVRATE15']              


# Replace null values with mean in the specified columns
for column in columns_to_fill:
    if column in df3.columns:
        mean_value = df3[column].mean()
        df3[column].fillna(mean_value, inplace=True) 

# Print the updated DataFrame
df3.isnull().sum()
```

```python
df3.to_excel('Food insecurity data3.xlsx')
```

```python
# List of column prefixes
column_prefixes = ['GROC', 'SUPERC', 'CONVS', 'SPECS', 'WICS', 'FFR']

# Function to estimate 2015 data for a given pair of 2011 and 2016 columns
def estimate_2015(x1, x2):
    # Calculate dx
    dx = (x2 - x1) / 5
    
    # Set a random seed for reproducibility
    np.random.seed(42)
    
    # Generate a random number between -dx and dx
    random_number = np.random.uniform(-dx, dx)
    
    # Estimate the 2015 data
    x3 = x2 - random_number
    
    return x3

# Create new columns for 2015 data
for prefix in column_prefixes:
    col_2011 = f'{prefix}11'
    col_2016 = f'{prefix}16'
    col_2015 = f'{prefix}15'
    
    df3[col_2015] = df3.apply(lambda row: estimate_2015(row[col_2011], row[col_2016]), axis=1)
```

```python
# List of column prefixes
column_prefixes = ['SNAPS']

# Function to estimate 2015 data for a given pair of 2012 and 2017 columns
def estimate_2015(x1, x2):
    # Calculate dx
    dx = (x2 - x1) / 5
    
    # Set a random seed for reproducibility
    np.random.seed(42)
    
    # Generate a random number between -dx and dx
    random_number = np.random.uniform(-dx, dx)
    
    # Estimate the 2015 data
    x3 = x2 - random_number
    
    return x3

# Create new columns for 2015 data
for prefix in column_prefixes:
    col_2012 = f'{prefix}12'
    col_2017 = f'{prefix}17'
    col_2015 = f'{prefix}15'
    
    df3[col_2015] = df3.apply(lambda row: estimate_2015(row[col_2012], row[col_2017]), axis=1)
```
